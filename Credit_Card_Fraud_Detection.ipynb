#CREDIT CARD FRAUD DETECTION 

```python
#importing the libraries
import pandas as pd
import numpy as np
```

```python
#importing the dataset
dataset = pd.read_csv('creditcard.csv')
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, -1].values
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
#distribution of legit transaction and fraudulent transaction
dataset['Class'].value_counts()
```

```python
#separating the data for analysis
legit = dataset[dataset.Class == 0]
fraud = dataset[dataset.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

STATISTICAL MEASURES OF THE DATA

```python
legit.Amount.describe()
```

```python
 fraud.Amount.describe()
```

```python
#comparing the values 
dataset.groupby('Class').mean()
```

UNDERSAMPLING:
BUILDING A SAMPLE DATASET

```python
legit_sample = legit.sample(n=492)
```

CONCATENATING TWO DATAFRAMES

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

SPLITTING DATA INTO TRAINING AND TESTING DATA

```python
#splitting data into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, stratify = Y, random_state=0)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

TRAINING THE MODEL ON LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, Y_train)
```

MODEL EVALUATION

```python
from sklearn.metrics import accuracy_score
X_train_prediction = log_reg.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(training_data_accuracy)
```

```python
X_test_prediction = log_reg.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_data_accuracy)
```
